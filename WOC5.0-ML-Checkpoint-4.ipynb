# Piechart


```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
plt.title("pie chart")
plt.tight_layout()
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0,0,0,0.1,0]


plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},explode=explode,shadow=True,startangle=180,autopct='%1.1f%%')
plt.show()
```

# Using Fill between


```python
import pandas as pd
from matplotlib import pyplot as plt
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv'
data = pd.read_csv(url)







ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='blue',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')



plt.legend()

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv'
data = pd.read_csv(url)






ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='blue',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')



plt.legend()
plt.fill_between(ages,py_salaries)
plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
url = 'https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv'
data = pd.read_csv(url)






ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='blue',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')
overall_median=57000



plt.legend()
plt.fill_between(ages,py_salaries,overall_median,interpolate=True,where=(py_salaries>overall_median),alpha=0.25,label='above avg')
plt.fill_between(ages,py_salaries,overall_median,interpolate=True,where=(py_salaries<=overall_median),alpha=0.25,label='below avg',color='green')

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
